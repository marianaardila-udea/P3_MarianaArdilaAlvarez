```python
import os
from clases import GestorDICOM

def main():
    gestor = GestorDICOM()
    while True:
        menu = input('''
        
        MENU SISTEMA
        1. Cargar carpeta DICOM
        2. Mostrar reconstrucci贸n 3D
        3. ZOOM
        4. Segmentaci贸n
        5. Transformaci贸n morfol贸gica
        6. Convertir a NIFTI
        7. Guardar CSV
        8. Salir
        ''')
        if menu == "1":
            ruta = input("Ruta carpeta (datos/(PPMI, T2, Sarcoma): ")
            gestor.cargar_carpeta(ruta)
        elif menu == "2":
            i = int(input(f"Estudio (0-{len(gestor.estudios)-1}): "))
            gestor.estudios[i].mostrar_reconstruccion_3d()
        elif menu == "3":
            i = int(input(f"Estudio: "))
            gestor.estudios[i].zoom()
            
        elif menu == "4":
            i = int(input(f"Estudio: "))
            gestor.estudios[i].segmentacion()
            
        elif menu == "5":
            i = int(input(f"Estudio: "))
            gestor.estudios[i].morfologica()
            
        elif menu == "6":
            i = int(input(f"Estudio: "))
            salida = input("Nombre .nii: ")
            gestor.convertir_nifti(i, salida)
        elif menu == "7":
            archivo = input("Nombre CSV (enter=estudios.csv): ") or "estudios.csv"
            gestor.guardar_csv(archivo)
            
        elif menu == "8":
            print("Saliendo del sistema...")
            break
        else:
            print("opcion no valida, intente nuevamente")

if __name__ == "__main__":
    main()
    
```
